```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series(["Sam",np.nan,"Tim","Kim"])
s
```

```python
s.isnull()
```

```python
s.notnull()
```

```python
s[3]=None
s.isnull()
```

```python
s.dropna()
```

```python
from numpy import nan as NA
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
df
```

```python
df[1]=NA
df
```

```python
df.dropna(axis=1,how="all")
```

```python
df
```

```python
df.dropna(thresh=3)
df
```

```python
df.fillna(0)
```

```python
df.fillna({0:15,1:25,2:35})
```

```python
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.fillna(method="ffill")
```

```python
df.fillna(method="ffill",limit=1)
```

```python
data=pd.Series([1,0,NA,5])
data
```

```python
data.fillna(data.mean())
```

```python
df
```

```python
df.fillna(df.mean())
```
