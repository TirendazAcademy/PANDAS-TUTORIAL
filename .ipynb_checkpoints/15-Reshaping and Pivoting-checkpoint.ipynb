```python
import pandas as pd
import numpy as np
```

```python
data=pd.DataFrame(
    np.arange(16).reshape(4,4),
    index=[list("aabb"),[1,2]*2],
    columns=[["num","num",
              "comp","comp"],
             ["math","stat"]*2])
data
```

```python
data.index.names=["class","exam"]
data.columns.names=["field","lesson"]
data
```

```python
long=data.stack()
```

```python
long
```

```python
long.unstack()
```

```python
data.stack()
```

```python
data.stack(0)
```

```python
data.stack("field")
```

```python
s1=pd.Series(
    np.arange(4),index=list("abcd"))
s2=pd.Series(
    np.arange(6,9),index=list("cde"))
```

```python
print(s1)
print(s2)
```

```python
data2=pd.concat([s1,s2],keys=["bir","iki"])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack(dropna=False)
```

```python
data2.unstack().stack(dropna=False)
```

## Pivoting ‚ÄúLong‚Äù to ‚ÄúWide‚Äù Format

```python
stock=pd.DataFrame(
    {"fruit": ["apple", "plum","grape"]*2,
     "color": ["purple","yellow"]*3,
     "piece":[3,4,5,6,1,2]})
```

```python
stock
```

```python
stock.pivot("fruit", "color", "piece")
```

```python
stock["value"]=np.random.randn(len(stock))
```

```python
stock
```

```python
p=stock.pivot("fruit","color")
p
```

```python
p["value"]
```

## Pivoting ‚ÄúWide‚Äù to ‚ÄúLong‚Äù Format

```python
data=pd.DataFrame(
    {"lesson":["math","stat","bio"],
     "Sam":[50,60,70],
     "Kim":[80,70,90],
     "Tom":[60,70,85]})
data
```

```python
group=pd.melt(data,["lesson"])
```

```python
group
```

```python
data=group.pivot(
    "lesson","variable","value")
data
```

```python
data.reset_index()
```
