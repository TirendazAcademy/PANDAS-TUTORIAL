```python
import pandas as pd
```

```python
film=pd.read_csv("http://bit.ly/imdbratings")
```

```python
film.head(3)
```

## Practical column selection

```python
film["title"].head() 
```

```python
film[["title","genre"]].head() 
```

## loc method

```python
film.loc[0,] 
```

```python
film.loc[[0,2,4],] 
```

```python
film.loc[0:2,] 
```

```python
film.loc[0:5,"title"]
```

```python
film.loc[0:5,"title":"genre"] 
```

```python
film.loc[0:5,"title":"duration"]
```

```python
film.loc[:,"title":"genre"].head()
```

```python
film.loc[film.genre=="Crime",].head(3) 
```

```python
film.loc[film.genre=="Crime",[
    "title","duration"]]
```

```python
film.loc[
    film.genre=="Crime","title":"duration"]
```

## iloc method

```python
film.iloc[:,0].head() 
```

```python
film.columns 
```

```python
film.iloc[:,[0,3]].head()
```

```python
film.iloc[:,0:3].head()
```

```python
film.iloc[0,0:3]
```

```python
film.iloc[0:5,0:3] 
```

```python
film.iloc[0:5,].head(3)
```

```python
film.iloc[0:5,:].head(3)
```

 ## Multiple filtering

```python
film.loc[film.duration>=200,].head(3)
```

```python
film.loc[film.duration>=200,"title"].head()
```

```python
film[(
    film.duration>=200)|(
    film.genre=="Crime")|(
    film.genre=="Action")].head(3)
```

```python
film[film.genre.isin([
    "Crime","Drama","Action"])]  
```
