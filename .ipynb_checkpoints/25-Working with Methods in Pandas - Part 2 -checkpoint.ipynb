```python
import pandas as pd  
import numpy as np
```

## to_datetime method

```python
pd.to_datetime("15/08/2019")
```

```python
date=["2019-01-05","jan 6, 2019",
      "7/05/2019","2019/01/9","20190110"]
```

```python
pd.to_datetime(date)
```

```python
pd.to_datetime("03/05/2019")
```

```python
pd.to_datetime("05/03/2019", dayfirst=True)
```

```python
pd.to_datetime("05*03*2019", 
               format="%d*%m*%Y" )
```

```python
pd.to_datetime("05$03$2019", 
               format="%d$%m$%Y")
```

```python
date=["2019-01-05",
     "jan 6, 2019",
     "7/05/2019",
     "2019/01/9",
     "20190110"]
```

```python
pd.to_datetime(date)
```

```python
pd.to_datetime(date, errors='coerce')
```

```python
t=1000000000
```

```python
pd.to_datetime(t, unit="s")
```

## Frequency and Date Offsets

```python
pd.date_range(
    "2010-01-01","2010-01-03", freq="4h")
```

```python
pd.date_range(
    "2010-01-01","2010-09-03", 
    freq="WOM-4SUN")
```

## Period and PeriodIndex

```python
p=pd.Period(2020)
p
```

```python
dir(p)
```

```python
p.start_time
```

```python
p.end_time
```

```python
a=pd.Period("2020-01", freq="M")
a
```

```python
a+5
```

```python
a-3
```

```python
p-pd.Period("2015")
```

```python
rng=pd.period_range(
    "2019-01-01","2019-08-30",freq="M")
rng
```

```python
pd.Series(range(8),index=rng)
```

```python
p=pd.Period("2019",freq="A-DEC")
p
```

```python
p.asfreq('M', how='start')
```

```python
p.asfreq("M",how="end")
```

```python
p=pd.Period("2019Q4", freq="Q-DEC")
```

```python
p=pd.Period("2019Q4", freq="Q-FEB")
p
```

```python
p.end_time
```

```python
p.asfreq('D', 'start')
```

```python
rng=pd.period_range(
    "2019Q3","2020Q4", freq="Q-JAN")
rng
```

```python
ts=pd.Series(range(len(rng)),index=rng)
ts
```

```python
rng=pd.date_range(
    "2020-01-01", periods=5, freq="M")
```

```python
ts=pd.Series(range(len(rng)),index=rng)
ts
```

```python
pts=ts.to_period()
pts.index
```

```python
pts.index
```
